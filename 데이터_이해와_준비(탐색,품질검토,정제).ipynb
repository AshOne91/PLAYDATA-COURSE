# 요약표만들기

````python
# 수치형 데이터
df = pd.read_csv(file_path,encoding='cp949')
numeric_summary = df.describe()
# 범주형 데이터
categorical_summary = df.describe(include='object')
# 결측치
missing_summary = df.isna().mean()
missing_summary = missing_summary[missing_summary>0]
````

````python
categorical_summary
````

````python
numeric_summary
````

````python
# 컬럼명이 달라서 옆으로 연결
summary_df =  pd.concat([numeric_summary,categorical_summary,missing_summary],axis=1)
summary_df
````

```
           count mean std min max unique top freq
 컬럼명1    00    nan              00
 컬럼명2    00    000                 
 컬럼명3    00
```

````python
summary_df = pd.DataFrame(index = df.columns,
             columns=['count','mean','std','min','max','unique','top','freq'])
# 각 컬럼에 대해서 통계를 채우기\
for col in df.columns:
  if df[col].dtype in ['int64','float64']: # 수치형
    desc = df[col].describe()
    summary_df.loc[col,'count'] = desc['count']
    summary_df.loc[col,'mean'] = desc['mean']
    summary_df.loc[col,'std'] = desc['std']
    summary_df.loc[col,'min'] = desc['min']
    summary_df.loc[col,'mas'] = desc['max']
  else: # 범주형
    desc = df[col].describe()
    summary_df.loc[col,'count'] = desc['count']
    summary_df.loc[col,'unique'] = desc['unique']
    summary_df.loc[col,'top'] = desc['top']
    summary_df.loc[col,'freq'] = desc['freq']
summary_df
````

````python
df = pd.read_csv(files[0])
# 상가업소번호와 코드가 들어간 명칭은 제외
new_cols = [col for col in df.columns if '코드' not in col]
new_cols = new_cols[1:11] + new_cols[-2:]
new_df = df[new_cols]
new_df.head(1)
````

````python
total_dfs = []
for file in tqdm(files):
  df = pd.read_csv(file,low_memory=False)
  # 상가업소번호와 코드가 들어간 명칭은 제외
  new_cols = [col for col in df.columns if '코드' not in col]
  new_cols = new_cols[1:11] + new_cols[-2:]
  new_df = df[new_cols]
  total_dfs.append(new_df)

# 리스트에 담기 데이터프레임들을 열(column) 기준으로 합친다  axis=1
merged_df = pd.concat(total_dfs,axis=0)
````

# 위경도를 알면 상점간의 거리를 알수 있다

````python
# 2D 형태... n행 x 2열
# 출력  (n, 2) 크기의 배열  n은 상점수
# n = 3
# [[000,000],
#  [000,000]
#  [000,000]]
import numpy as np
coords = df[['경도','위도']].to_numpy()
print(coords.shape)
# 차원 확장  (n, 2) --> (n,1,2)  (3,2) --> (3,1,2)
coords2 = np.expand_dims(coords,axis=1)
print(coords2.shape)  # (553535, 1, 2)
# (553535, 1, 2) 배열과 (553535,  2) 간의 차이를 계산  모든 상점 쌍간의 좌표 차이를 구함
# (n,n,2)  -->(경도차이, 위도차이)
````

````python
import pandas as pd
import numpy as np
from glob import glob
from tqdm import tqdm  # 진행형황
files = glob('/content/drive/MyDrive/data/소상공인/sub1/*.csv')
files[6]
df = pd.read_csv(files[6])
````

````python
# 표준 편차 : 각 값들이  평균으로부터 얼마나 떨어져 있는지 나타내는 지표
summary_status =  df.groupby(['시군구명']).agg({
    '상호명' : 'count',  # 상점 수
    '경도' : ['mean','std'],
    '위도' : ['mean','std']
}).round(2)
summary_status.columns = ['상점 수','평균_경도','경도_표준편차','평균_위도','위도_표준편차']
for category in ['음식','소매']:
  summary_status[f'{category} 업종비율 (%)']  = \
    (df[df['상권업종대분류명'] == category].groupby('시군구명').size() / summary_status['상점 수'] * 100).round(2)

summary_status
````
